```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Iterasi maksimum
maxIter = 500

lenX = lenY = 20  # Bujur sangkar
delta = 1

# Kondisi batas
Ttop = 100
Tbottom = 0
Tleft = 0
Tright = 0

# grid dalam
Tguess = 30
```

```python
colorinterpolation = 50
colourMap = plt.cm.jet

# kisi-kisi
X, Y = np.meshgrid(np.arange(0, lenX), np.arange(0, lenY))
```

```python
T = np.empty((lenX, lenY))
T.fill(Tguess)

# Batas batas
T[(lenY-1):, :] = Ttop
T[:1, :] = Tbottom
T[:, (lenX-1):] = Tright
T[:, :1] = Tleft
```

```python
# Iterasi (asumsikan bahwa iterasi adalah konvergensi di maxIter = 500)
print("Mohon Tunggu. . . . .")
for iteration in range(0, maxIter):
    for i in range(1, lenX-1, delta):
        for j in range(1, lenY-1, delta):
            T[i, j] = 0.25 * (T[i+1][j] + T[i-1][j] + T[i][j+1] + T[i][j-1])

print("Iterasi Selesai")
```

```python
# Konfigurasi
plt.title("Kontur Suhur")
plt.contourf(X, Y, T, colorinterpolation, cmap=colourMap)
plt.colorbar()
plt.show()

print("")
```
